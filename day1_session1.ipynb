## Sentiment analysis using transformer models

```python
from transformers import pipeline

classifier = pipeline("sentiment-analysis")
classifier(["I've been waiting for a HuggingFace course my whole life.",
            "My Noodles are spicy.",
            "Nicky lost his mind.",
            "Superman was Fire.",
            "Joker was shit."])
```

## Text generation model(like Chatgpt)

```python
generator = pipeline("text-generation")
generator("Once upon a time in National college of Engineering..")
```

```python
qa = pipeline("question-answering")
qa(
    question = "Where do I work?",
    context = "Once upon a time in National College of Engineering."
)
```

# Language Translator


```python
translator = pipeline("translation_en_to_de")
translator("I love myself")
```

## NER Models

```python
ner  = pipeline("ner", model = "dbmdz/bert-large-cased-finetuned-conll03-english")
ner("I am Aries and I am from Kirtipur Nepal")
```
